```python
""" pathlib 模块 """

from pathlib import Path

## 查看当前工作路径
print(Path.cwd())
p = Path("e:\ProjectDocumentStorage\python\pythonStudy\src")
print(p)
q = p / "data\Answer.txt"
## 判断是否为文件夹
print(p.is_dir())
print(q.is_dir())
## 判断是否为文件
print(p.is_file())
print(q.is_file())
## 判断路径是否存在
print(p.exists())
print(q.exists())
## 获取路径的最后一部分
print(p.name)
print(q.name)
## 获取文件名
print(q.stem)
## 获取文件后缀
print(q.suffix)
## 获取父级目录
print(p.parent)
print(q.parent)
## 获取到硬盘根目录的所有父级目录
ps = p.parents
for each in ps:
    print(each)

## 将路径拆分为元组
print(p.parts)

## 查询文件或文件夹的信息
print(p.stat())
print(q.stat())

## 将相对路径转化为绝对路径
print(Path("./data").resolve())

## 获取子文件夹/文件路径
it = p.iterdir()
for each in it:
    print(each)

# 将文件路径整理为列表
print([x for x in p.iterdir() if x.is_file()])

## 创建文件夹
n = p / "temp"
n.mkdir(exist_ok=True)

## 创建多级文件夹
ns = p / "temp/A/B/C"
ns.mkdir(parents=True, exist_ok=True)
## 修改文件/文件夹名称
# q.rename("Answer6.txt")
## 删除文件
# ns.unlink()
## 删除文件夹
# ns.parent.rmdir()

## 查找该目录下文件
g = p.glob("*.ipynb")
print([each for each in g])
## 查找该目录及子目录下的文件
g = p.glob("**/*.ipynb")
print([each for each in g])
```

```python
""" 异常 exception """

## 捕获处理异常
try:
    1 / 0
except ZeroDivisionError as e:
    print("出错了~", e)

try:
    1 / 0
    520 + ""
except ZeroDivisionError:
    print("除数不能为0!")
except ValueError:
    print("值不正确！")
except TypeError:
    print("类型不正确！")

## try-except-else
try:
    1 / 0
except ZeroDivisionError:
    print("逮到了！")
else:
    print("没逮到！")

## try-except-finally
try:
    1 / 0
except ZeroDivisionError:
    print("逮到了！")
else:
    print("没逮到！")
finally:
    print("吱一声！")

## raise 语句 直接抛出异常
# raise ValueError("值错误！")

## 异常链 raise from
# raise ValueError("这样不行~") from ZeroDivisionError

## 代码调试 assert boolean
assert "hello" == "hello"

## 利用异常实现 goto
try:
    while 1:
        while 1:
            for i in range(10):
                if i > 3:
                    raise
                print(i)
            print("被跳过~")
        print("被跳过~")
    print("被跳过~")
except:
    print("到这来了！")
```
