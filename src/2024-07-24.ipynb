```python
import random

""" 用 Python 设计第一个游戏（改进） """
answer = random.randint(1, 10)
while 1:
    temp = input("猜一下数字")
    guess = int(temp)
    if guess == answer:
        print("猜中了")
        break
    else:
        if guess < answer:
            print("小了")
        else:
            print("大了")
print("结束")
```

## 格式化字符串
`[[fill]align][sign][#][0][width][grouping_option][.precision][type]`
### align
- '<'：强制字符串左对齐
- '>'：强制字符串右对齐
- '='：强制填充0在符号和数字之间
- '^'：强制字符串居中
### sign
- '+'：在正数前面加（+），在负数前面加（-）
- '-'：在负数前面加（-），默认
- 空格：正数前面加空格，负数前面加（-）
### type
- 'b'：将参数转化为二进制
- 'c'：将参数以 Unicode 字符输出
- 'd'：将参数进行十进制输出
- 'o'：将参数进行八进制输出
- 'x'：将参数进行十六进制输出
- 'X'：将参数进行十六进制输出
- 'n'：跟'd'类似，不同之处在于他会使用当前语言环境设置的分隔符插入到恰当的位置
- None：跟'd'一样
- 'e'：将参数进行科学计数法输出（以字母e作为指数，默认精度为6）
- 'E'：将参数进行科学计数法输出（以字母E作为指数，默认精度为6）
- 'f'：将参数进行定点表示法输出（不是数字用'nan'，无穷用'inf'表示，默认精度为6）
- 'F'：将参数进行定点表示法输出（不是数字用'NAN'，无穷用'INF'表示，默认精度为6）
- 'g'：通用格式，小数以'f'形式进行输出，大数以'e'进行输出
- 'G'：通用格式，小数以'F'形式进行输出，大数以'E'进行输出
- 'n'：跟'g'类似，不同之处在于他会使用当前语言环境设置的分隔符插入到恰当的位置
- '%'：以百分比格式输出，小数点后最少会有一位

```python
## 格式化字符串
year = 2024
print("年：{}".format(year))
month = 12
day = 15
hour = 12
minute = 30
second = 45
print("{0}-{1}-{2} {3}:{4}:{5}".format(year, month, day, hour, minute, second))

## 输出花括号
print("{},{{}},{}".format(1, 2))

## 格式化
print("{:^10}".format(250))

print("{1:>10}{0:<10}".format(520, 250))

print("{:010}".format(250))
print("{:010}".format(-250))
print("{1:%>10}{0:%<10}".format(520, 250))
print("{:0=10}".format(-250))

## sign
print("{:+}{:-}".format(520, -250))
print("{:,}".format(12345678))
print("{:_}".format(12345678))
## .precision
## 保留两位小数
print("{:.2f}".format(3.1415926))
## 两位有效数字
print("{:.2g}".format(3.1415926))
## 非数字进行截取
print("{:.6}".format("I love Python"))

## type
print("{:#b}".format(80))
print("{:c}".format(80))
print("{:d}".format(80))
print("{:#o}".format(80))
print("{:#x}".format(80))
print("{:}".format(80))
print("{:e}".format(3.1415))
print("{:E}".format(3.1415))
print("{:g}".format(123456789))
print("{:.{prey}%}".format(0.9808, prey=2))

## 语法糖 f
print(f"今年是{year}")
print(f"{-520:010}")
fill = "+"
align = "^"
width = 10
prey = 3
ty = "g"
print(f"{3.1415:{fill}{align}{width}.{prey}{ty}}")
```
