```python
"""描述符"""
class D:
    def __get__(self,instance,owner):
        print(f"get~\nself -> {self}\ninstance -> {instance}\nowner -> {owner}")
    def __set__(self,instance, value):
        print(f"set~\nself -> {self}\ninstance -> {instance}\nvalue -> {value}")
    def __delete__(self,instance):
        print(f"delete~\nself -> {self}\ninstance -> {instance}")

class C:
    x = D()

c = C()
c.x = 250

c.x

class D:
    def __get__(self,instance,owner):
        return instance._x
    def __set__(self,instance,value):
        instance._x = value
    def __delete_(self,instance):
        del instance._x
class C:
    def __init__(self,x = 250):
        self._x = x
    x = D()

c = C()
c.x
c.x = 520

## 利用描述符造出 property() 函数
class MyProperty():
    def __init__(self,fGet= None,fSet=None,fDel=None):
        self.fGet = fGet
        self.fSet = fSet
        self.fDel = fDel
    def __get__(self,instance,owner):
        return self.fGet(instance)
    def __set__(self,instance,value):
        return self.fSet(instance,value)
    def __selete__(self,instance):
        return self.fDel(instance)
    def getter(self,func):
        self.fGet = func
        return self
    def setter(self,func):
        self.fSet = func
        return self
    def deleter(self,func):
        self.fDel = func
        return self

class C:
    def __init__(self):
        self._x = 250
    def getX(self):
        return self._x
    def setX(self,value):
        self._x = value
    def delX(self):
        del self._x
    x = MyProperty(getX, setX, delX)

c = C()
print(c.x)

class D:
    def __init__(self):
        self._x = 250
    @property
    def x(self):
        return self._x
    @x.setter
    def x(self,value):
        self._x = value
    @x.deleter
    def x(self):
        del self._x
d = D()

print(d.x)
d.x = 520
print(d.__dict__)
##todo 描述符只能运用于类属性，运用到方法属性则无效
```

```python
"""元类"""
from typing import Any


class MetaC(type):
    pass

class C(metaclass=MetaC):
    pass

c= C()
print(type(c))
print(type(MetaC))

class MetaC(type):
    def __new__(mcls,name,bases,attrs):
        print("__new__() in MetaC~")
        print(f"mcls={mcls}, name={name}, bases={bases}, attrs={attrs}")
        return type.__new__(mcls,name,bases,attrs)
    def __init__(cls,name,bases,attrs):
        print("__init__() in MetaC~")
        print(f"mcls={cls}, name={name}, bases={bases}, attrs={attrs}")
        return type.__init__(cls,name,bases,attrs)

class C(metaclass=MetaC):
    def __new__(cls):
        print("__new__() in C~")
        return super().__new__(cls)
    def __init__(self):
        print("__init__() in C~")

c = C()

class MetaC(type):
    def __call__(self, *args: Any, **kwds: Any):
        print("__call__() in MetaC~")

class C(metaclass=MetaC):
    pass

c = C()
```

```python
"""元类实际开发中运用"""
class MetaC(type):
    def __new__(mcls, name, bases, attrs):
        attrs["author"] = "FishC"
        return type.__new__(mcls,name,bases,attrs)
    
    def __init__(cls, name:str, bases, attrs):
        ## 判断类名必须为大写字母开头
        if not name.istitle():
            raise TypeError("类名必须为大写字母开头~")
        cls.createDate = "2024-07-05"
        return type.__init__(cls,name,bases,attrs)
    def __call__(cls, *args, **kwargs):
        if args:
            raise TypeError("仅支持关键字参数")
        ## 将实例的属性值大写
        new_args = [each.upper() for each in args if isinstance(each, str)]
        return type.__call__(cls,*new_args,**kwargs)

class C(metaclass=MetaC):
    pass

class D(metaclass=MetaC):
    pass

#! class mycls(metaclass=MetaC):
class Mycls(metaclass=MetaC):
    pass

class Upperproperty(metaclass= MetaC):
    def __init__(self, name):
        self.name = name

c = C()
d = D()
print(c.author)
print(d.author)
print(c.createDate)
print(d.createDate)

upperProperty = Upperproperty(name="FishC")
print(upperProperty.name)

class NoInstances(type):
    def __call__(cls,*args,**kwargs):
        raise TypeError("该类不允许直接实例化对象!")

class C(metaclass=NoInstances):
    @staticmethod
    def static_ok():
        print("可以允许直接访问静态方法~")
    @classmethod
    def class_ok(cls):
        print("类方法可以直接访问")
        
class SimpleInstances(type):
    def __init__(cls,*args,**kwargs):
        cls.__instance = None
        type.__init__(cls,*args,*kwargs)
    def __call__(cls,*args,**kwargs):
        if cls.__instance is None:
            cls.__instance = type.__call__(cls,*args,**kwargs)
            return cls.__instance
        else:
            return cls.__instance

#! c = C()
C.static_ok()
C.class_ok()

class C(metaclass=SimpleInstances):
    pass

c1 = C()
c2 = C()
print(c1 is c2)
```

```python
"""抽象基类"""
from abc import ABCMeta,abstractmethod
class Fruit(metaclass=ABCMeta):
    def __init__(self,name):
        self.name = name
    @abstractmethod
    def good_for_health(self):
        pass

#! fruit = Fruit("水果") ## 不能直接实例化

class Banana(Fruit):
    def good_for_health(self):
        print("只要吃香蕉，人就会变的开心~")
banana = Banana("香蕉")
banana.good_for_health()
```
